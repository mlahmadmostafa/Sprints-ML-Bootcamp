```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(0, 10, 40)
```

# Basic Line and Scatter Plots

### Add a line plot

```python
# scatter plot
plt.figure(figsize=(6,6))
plt.plot(x, x*.75)
plt.title("Line plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.style.use("ggplot")
```

### Add a scatter plot

```python
# scatter plot
plt.figure(figsize=(6,6))
plt.scatter(x, x*.75)
plt.title("Scatter plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.style.use("ggplot")
```

# Advanced Plotting

### Bar:

```python
x = np.linspace(0,100,20)
y = np.random.randint(0,100,20)
fig, ax = plt.subplots(1,2,figsize = (12,6))
ax[0].bar(x,y, edgecolor='black', color = "hotpink")
ax[0].set_title("Bar Plot")
ax[0].grid()

ax[1].hist(y,bins = 5, edgecolor='black', color = "gold")
ax[1].set_title("Histogram of y")

plt.tight_layout()
```

# 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(0,5,20) # Creating x array
y = np.linspace(0,5,20) # Creating y array
x, y = np.meshgrid(x, y) # Adding 2d grid
z = np.sin(np.sqrt(x**(-1/2) + y**(-1/2)))


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap="rainbow", edgecolor='k') # Plotting
```
