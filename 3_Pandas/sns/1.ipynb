```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Line plot 

```python
sns.set_theme(style="whitegrid")
x = np.linspace(0, 10, 50)
y_line = np.sin(x)

plt.figure(figsize=(10, 5))
sns.lineplot(x=x, y = y_line, marker='o',  label='sin(x)')
plt.title("Line Plot: sin(x)", fontsize=14)
plt.xlabel("X-axis",)
plt.ylabel("Y-axis", fontsize=12)
```

# Scatter plot

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=x, y = y_line, marker='o',  label='sin(x)')
plt.title("Scatter Plot: sin(x)", fontsize=14)
plt.xlabel("X-axis",)
plt.ylabel("Y-axis", fontsize=12)
```

# Box plot

```python
import pandas as pd

np.random.seed(42)
data = pd.DataFrame({
    'Numerical': np.random.normal(loc=50, scale=10, size=100),
    'Category': np.random.choice(['A', 'B', 'C'], size=100, replace=True)
})
sns.boxplot(x='Category', y='Numerical', data=data)
plt.title("Box Plot of Numerical by Category")
plt.xlabel("Category")
plt.ylabel("Numerical")
```

# Violin plot

```python
sns.violinplot(x='Category', y='Numerical', data=data)
plt.title("Violin Plot of Numerical by Category")
plt.xlabel("Category",)
plt.ylabel("Numerical",)
```

# Heatmap

```python
data = pd.DataFrame({
    'Feature_1': np.random.rand(100),
    'Feature_2': np.random.rand(100),
    'Feature_3': np.random.rand(100),
    'Feature_4': np.random.rand(100),
})

correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix, 
    annot=True,  # Annotate with correlation values
    fmt=".2f",   # Format for the annotations
    cmap="coolwarm",  # Color map
    linewidths=0.5,   # Add lines between cells
    cbar_kws={'label': 'Correlation Coefficient'}  # Label for the color bar
)
plt.title("Correlation Matrix Heatmap")
```
